## 2.5.2 Python第三方库的安装与使用

```python
import random
print(random.randint(1, 8))
from numpy import random
print(random.randint(1,8,(3,2)))
print(random.randint(1,8))
```

```python
import numpy as np
print(np.random.randint(1,8,(3,2)))      #在1-8的整数中随机创建3行2列的数组
```

```python
import pandas as pd
print(pd.Series((2,3,6)))         #从list（2,3,6）创建Series
```
